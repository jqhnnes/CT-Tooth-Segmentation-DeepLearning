### Transfer Data names into nnUNetv2 Format

```python
import os

def rename_files_with_0000(input_folder):
    for filename in os.listdir(input_folder):
        if filename.endswith(".nii.gz") and not filename.endswith("_0000.nii.gz"):
            base_name = filename[:-7]  # Entferne ".nii.gz"
            new_name = base_name + "_0000.nii.gz"
            old_path = os.path.join(input_folder, filename)
            new_path = os.path.join(input_folder, new_name)
            print(f"Rename: {filename} -> {new_name}")
            os.rename(old_path, new_path)

# Beispiel Pfad anpassen
input_folder = "/ssd/geiger/Segmentierung-von-CT-Aufnahmen-extrahierter-Z-hne-mittels-Deep-Learning/data/nnUNet_raw/Dataset001_GroundTruth/imagesTr"
rename_files_with_0000(input_folder)
```
