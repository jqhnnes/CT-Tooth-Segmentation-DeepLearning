### Function Definition: Check and Display Unique Labels in NIfTI Files

This function loads a specified number of `.nii.gz` label files from a given folder and prints all unique label values found in each file. It is useful for quickly inspecting segmentation label consistency or verifying that the expected label values are present across samples.

It supports both relative and absolute paths, ensuring reliable file access regardless of the execution environment.

**Parameters:**
- `labels_folder` (str): Path to the folder containing the label `.nii.gz` files.  
- `n_samples` (int): Number of label files to inspect.

**Behavior:**
- Loads up to `n_samples` label files using the `nibabel` library.  
- Extracts and prints unique label values found in each file using `numpy`.  
- Displays label statistics for each selected file in a clear and readable format.

```python
import nibabel as nib
import numpy as np
import os

def check_labels_and_display(labels_folder, n_samples):
    """
    Load and check a specified number of label files, printing the unique labels present in each.

    Parameters
    ----------
    labels_folder : str
        Path to the folder containing label .nii.gz files.
    n_samples : int
        Number of label files to inspect.

    Returns
    -------
    None
    """
    label_files = [f for f in os.listdir(labels_folder) if f.endswith(".nii.gz")]
    sampled_files = label_files[:n_samples]

    for file in sampled_files:
        path = os.path.join(labels_folder, file)
        img = nib.load(path)
        data = img.get_fdata()
        unique_labels = np.unique(data)
        print(f"Labels present in {file}: {unique_labels}")
```

### Function Execution

The following function call runs the label inspection process on the first 10 `.nii.gz` label files found in the specified directory. It prints each file name along with the set of unique label values detected within it, allowing for a quick check of segmentation consistency and labeling correctness.

```python
check_labels_and_display("../data/nnUNet_raw/Dataset001_GroundTruth/imagesTr", 10)
```
